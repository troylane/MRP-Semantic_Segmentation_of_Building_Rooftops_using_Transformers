```python
def move_images(image_list):
    src_folder = r"D:/courses/MRP/dataset/final/annotations/test/"
    for names in image_list:
        dst_folder = r"D:/courses/MRP/dataset/final/annotations/move_tiles_per_folder/"+names+"/"
        print(dst_folder)
        pattern = names+"_*.png"
        files = glob.glob(src_folder + pattern)
        for file in files:
            # extract file name form file path
            file_name = os.path.basename(file)
            shutil.move(file, dst_folder + file_name)
            print('Moved:', file)
            
        
```
